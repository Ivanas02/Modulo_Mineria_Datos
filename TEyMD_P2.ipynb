## *Estadística descriptiva*

Se ocuparán las siguientes estadísticas a cada uno de los campos:

- *Media:* Valor promedio de un conjunto de datos numéricos.
- *Mínimo*: Es el valor más pequeño que toma una función/campo.
- *Máximo:* Es el valor más grande que toma una función/campo.
- *Desviación estándar:* Medida que indica qué tan dispersos están los datos en relación a la media
- *Conteo/Frecuencia:* Cantidad de valores no nulos dentro de un campo.

```python
# Eliminamos la variable categorica
df_est = data.drop('c',axis=1)
# Aplicamos la estadística descriptiva
df_est = pd.DataFrame({
    'Media': df_est.mean(),
    'Mínimo': df_est.min(),
    'Máximo': df_est.max(),
    'Desviación Estándar': df_est.std(),
    'Conteo': df_est.count()
})
df_est
```

## *Visualización de datos*

- Gráfico de barras (género de encuestados)

Un gráfico de barras nos sirve para identificar la frecuencia asociada a cada uno de los distintos valores que toma un campo dentro de una base de datos.

En este caso visualizaremos la distribución por género de los que se encuentran dentro de la base.

```python
import matplotlib.pyplot as plt

# Contar las frecuencias de cada categoría en la columna 'g'
counts = df['g'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
counts.plot(kind='bar', color='skyblue')
plt.title('Distribución por sexo')
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Normalmente en una encuesta se secciona el género de los participantes en 3 *masculino, femenino y otro*, para fines prácticos tomaremos *1: masculino*, *2: femenino* y *3: otro*, dentro del campo 3 agruparemos todos aquellos valores que son diferentes de 1 o 2.

```python
df['g'] = df['g'].apply(lambda x: x if x in [1, 2] else 3)
```

```python
# Contar las frecuencias de cada categoría en la columna 'g'
counts = df['g'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
counts.plot(kind='bar', color='skyblue')
plt.title('Distribución por sexo')
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

- Gráfica de barras (valoración por sexo sobre **N10**)

Es la misma función del gráfico solo que para este ocupamos distintos insumos, dado que queremos conocer la distribución de la valoración de obtenida dentro de la columna *N10* teniendo en cuenta los distintos géneros de los encuestados.

```python
# Contar las valoraciones por sexo
df_grouped = df.groupby(['N10', 'g']).size().unstack().fillna(0)

# Crear el gráfico de barras
ax = df_grouped.plot(kind='bar', figsize=(10, 6), colormap='viridis')

# Personalizar el gráfico
plt.title('Distribución de Valoraciones por Sexo')
plt.xlabel('Valoración')
plt.ylabel('Número de Respuestas')
plt.legend(title='Sexo')
plt.xticks(rotation=0)
plt.show()
```

- Gráfica de barras (valoración por sexo sobre **O10**)

Es la misma función del gráfico solo que para este ocupamos distintos insumos, dado que queremos conocer la distribución de la valoración de obtenida dentro de la columna *O10* teniendo en cuenta los distintos géneros de los encuestados.

```python
# Contar las valoraciones por sexo
df_grouped = df.groupby(['O10', 'g']).size().unstack().fillna(0)

# Crear el gráfico de barras
ax = df_grouped.plot(kind='bar', figsize=(10, 6), colormap='viridis')

# Personalizar el gráfico
plt.title('Distribución de Valoraciones por Sexo')
plt.xlabel('Valoración')
plt.ylabel('Número de Respuestas')
plt.legend(title='Sexo')
plt.xticks(rotation=0)
plt.show()
```

- Gráfico de líneas

Para este gráfico ocuparemos la edad de los encuestados para poder agruparlos y de esa manera saber la participación que se tuvo dependiendo de la edad de estos mismos.

Como lo mencionamos anteriormente la base viene con algunas deficiencias (son datos en crudo), en algunos casos se colocó el año de nacimiento en lugar de la edad, así que calcularemos la edad suponiendo que la encuesta se realizó en este año en curso.

```python
from datetime import datetime
anio = datetime.now().year
```

Para todos aquellos que tienen un valor mayor a *1000*, realizamos la diferencia entre el año actual y su fecha de nacimiento.

```python
df['a'] = df['a'].apply(lambda x: x if x < 1000 else anio - x)
```

```python
df_res = df.groupby('a').agg(
    frecuencia=('a', 'count')
).reset_index()
df_res = df_res[df_res['a']>0]
df_res = df_res[df_res['a']<100]
```

```python
# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(df_res['a'], df_res['frecuencia'], marker='o', linestyle='-', color='skyblue')

# Personalizar el gráfico
plt.title('Frecuencia por edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()  # Ajusta el diseño para que no se corten las etiquetas
plt.show()
```

Se puede realizar una representación similar del gráfico anterior teniendo como base un gráfico de barras, pero al ser demasiados valores en el eje x se volvería un gráfico nada atractivo visualmente.

Para ello realizamos la creación de una nueva columna, agrupando por rangos para poder tener una mejor representación e interpretación de lo que los datos nos estan diciendo.

```python
# Definir los límites de los rangos de edad
bins = [13, 18, 30, 40, 65, 100]
labels = ['Menor de 18', '18-30', '30-40', '40-65', '65-100']

# Crear una nueva columna con los rangos de edad
df['rango_edad'] = pd.cut(df['a'], bins=bins, labels=labels, right=False)

df.head()
```

Agrupamos los registros conforme a la columna anteriormente creada, de esta manera obtendremos la frecuencia asociada a cada rango.

```python
df_res = df.groupby('rango_edad').agg(
    frecuencia=('a', 'count')
).reset_index()
df_res
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_res['rango_edad'],df_res['frecuencia'], color='darkblue')
plt.title('Distribución por edad')
plt.xlabel('Rango')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

- Gráfico de burbuja (con respecto a la columna **N10**)

Un gráfico de burbujas es una variación de un gráfico de dispersión en el que los puntos de datos se reemplazan por burbujas y se representa una dimensión adicional de los datos en el tamaño de las burbujas. Al igual que un gráfico de dispersión, un gráfico de burbujas no usa un eje de categorías; los ejes horizontal y vertical son ejes de valores. Además de los valores x e y que se trazan en un gráfico de dispersión, un gráfico de burbujas traza valores x, valores y y valores z (tamaño).

Para obtener este gráfico, primero agrupamos los datos por edad calculando el promedio asociado con respecto a los datos de la columna *N10*, además obtenemos la frecuencia de estos mismos.

```python
df_res = df.groupby('a').agg(
    promedio_calificacion=('N10', 'mean'),
    frecuencia=('a', 'count')
).reset_index()
df_res = df_res[df_res['a']>0]
df_res = df_res[df_res['a']<100]
```

```python
# Crear el gráfico de burbuja
plt.figure(figsize=(10, 6))
plt.scatter(
    df_res['a'],
    df_res['promedio_calificacion'],
    s=df_res['frecuencia'],  # Tamaño de las burbujas
    alpha=0.5,
    color='skyblue',
    edgecolors='w',
    linewidth=0.5
)
plt.title('Niveles de Satisfacción por Intervalo de Edad')
plt.xlabel('Valores')
plt.ylabel('Nivel de Satisfacción')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)
plt.tight_layout()
plt.show()
```

El tamaño nos indica la frecuencia obtenida, es decir, a mayor números de registros asociados a cierta edad la burbuja que lo representa será más grande.

Podemos realizar un ejercicio similar pero ahora teniendo un panorama general por rangos, en este caso ocuparemos la columna antes creada, calculando el promedio de los valores asociados a la columna **O10** y representandolos en un gráfico de burbujas.

```python
df_res = df.groupby('rango_edad').agg(
    promedio_calificacion=('O10', 'mean'),
    frecuencia=('a', 'count')
).reset_index()
df_res
```

```python
# Crear el gráfico de burbuja
plt.figure(figsize=(10, 6))
plt.scatter(
    df_res['rango_edad'],
    df_res['promedio_calificacion'],
    s=df_res['frecuencia'],  # Tamaño de las burbujas
    alpha=0.5,
    color='skyblue',
    edgecolors='w',
    linewidth=0.5
)
plt.title('Niveles de Satisfacción por Intervalo de Edad')
plt.xlabel('Rango')
plt.ylabel('Nivel de Satisfacción')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)
plt.tight_layout()
plt.show()
```

## *Transformación para valores faltantes o datos inconsistentes*

Para manejar valores faltantes o datos inconsistentes en un conjunto de datos, se pueden utilizar técnicas de limpieza de datos, como:

- **Imputación:** Reemplazar los valores faltantes con estimaciones basadas en los datos presentes. Para ello, se puede utilizar la media, la mediana o la moda de la variable.
- **Eliminación de datos:** Eliminar los registros que contengan datos faltantes o las columnas que tengan valores faltantes. Esta técnica puede simplificar el conjunto de datos, pero puede resultar en la pérdida de información.
- **Detección de valores atípicos:** Identificar valores inusuales que se desvían de la mayoría de los datos. Estos valores pueden ser corregidos o excluidos.
- **Consistencia de datos:** Garantizar que los datos sean coherentes en términos de formato y unidades.

Dentro de la elaboración de esta práctica se llevaron dos transformaciones de datos, la primera fue con respecto al género (columna g), dado que había pocos valores qeu no hacían sentido, así que se decidió agruparlos para asociarlos a un mismo valor y que de esta manera representaran el mismo significado.

El segundo fue durante la transformación/limpieza de los valores con respecto a la edad (columna a), aqui se interpretó que se estaban asociando años en lugar de edades, por lo que se realizó la estimación de la edad tomando como base el año actual para poder determinar su longevidad. Dentro de este mismo campo, aquellos valores que no nos hacían coherencia no se tomaron en cuenta para realizar las representaciones graficas.

## *Matriz de correlación*

Enlistamos las columnas seleccionadas

```python
lista = ['a','A10','C10','E10','N8','N7','N10','O5','O10','C10',]
df_matriz = data[lista]
```

```python
matriz_correlacion = df_matriz.corr()

# Visualizar la matriz de correlación
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlación de 10 Columnas Seleccionadas')
plt.show()
```

Con respecto a la matriz de correlación anteriormente generada, podemos observar que se presenta una correlación "positiva" entre *N7-N8* y *O5-O10*, es decir cuando una variable aumenta la otra tambien tiende a aumentar. El mayor número de correlaciones se identifican como nulas, esto significa que los cambios en una variable no pueden predecir los cambios en la otra.
